```python
import pandas as pd 
import geopandas as gpd
```

```python
# Load and read data 
neighborhoodJSON = './dallas_coordinates.json'
housesCSV = './houses.csv'

houses = pd.read_csv(housesCSV)
neighborhoods = gpd.read_file(neighborhoodJSON)
```

```python
# convert lat/lng data to geometry with gpd 
gdp_df = gpd.GeoDataFrame(
    houses,   geometry=gpd.points_from_xy(houses.lng, houses.lat))
gdp_df
```

```python
# join with shape file 
join = gpd.sjoin(gdp_df, neighborhoods)
join
```

```python
# Clean 
join = join.drop(columns=['Unnamed: 0','index_right', 'cartodb_id', 'created_at', 'updated_at'])
```

```python
join.to_csv('./joined.csv')
```
